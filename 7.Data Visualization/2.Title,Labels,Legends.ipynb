```python
import matplotlib.pyplot as plt
```

```python
x_axis=[i for i in range(1,11)]
y_axis=[i**2 for i in range(1,11)]
```

```python
#title
plt.title("Mihir's Graph")
plt.plot(x_axis,y_axis,'red')
```

```python
#label
plt.xlabel("Numbers")
plt.ylabel("Squares")
plt.plot(x_axis,y_axis,"r")
```

```python
x2_axis=[i for i in range(1,11)]
y2_axis=[i**3 for i in range(1,11)]
```

```python
#legend
plt.title("Two Plots in one graph")
plt.xlabel("Numbers")
plt.ylabel("Square and Cubes")
plt.plot(x_axis,y_axis,'r')
plt.plot(x2_axis,y2_axis)
plt.legend(["Square","Cube"])
```
