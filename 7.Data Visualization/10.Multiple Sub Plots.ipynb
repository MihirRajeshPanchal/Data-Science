```python
import matplotlib.pyplot as plt
import random as r
```

```python
l1=[i for i in range(1,11)]
l2=[]
l3=[i for i in range(11,21)]
l4=[]
for i in range(1,11):
    l2.append(r.randint(1,10))
    l4.append(r.randint(11,20))
```

```python
plt.subplot(1,2,1)
plt.title("Red")
plt.xlabel("1-10 X")
plt.ylabel("1-10 Y")
plt.plot(l1,l2,'r')
plt.subplot(1,2,2)
plt.title("Green")
plt.xlabel("11-20 X")
plt.ylabel("11-20 Y")
plt.plot(l3,l4,'g')
plt.tight_layout()
```
