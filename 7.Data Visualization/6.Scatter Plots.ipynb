```python
import matplotlib.pyplot as plt
import pandas as pd
import random as r
```

```python
l1=[i for i in range(1,11)]
scatter=[]
for i in range(1,11):
    scatter.append(r.randrange(1,100))
df=pd.DataFrame([l1,scatter])
df.set_axis(["Numbers","Scatter Y"])
```

```python
plt.scatter(l1,scatter,c=l1,cmap="Accent")#,color='r')
```
